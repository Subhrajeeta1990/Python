```python
#WAP if number is ending with 5 0r 0, break the loop, print the number. 
#if it not, divide it with 3 and check if the result is div by 2, if yes break loop, if no continue.

while True:
    L=[1,23,45,64,38,95]
    for i in L:
        i=str(int(i))
        if " 5" in i or " 0" in i:
            print(i)
            break
        else:
            i=int(str(i))
            while i in L:
                a=i/3
                print(a)
                if a%2==0:
                    print("Yes")
                    break
                elif a%2!=0:
                    print("No")
                    continue
```
